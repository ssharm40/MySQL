# Trigger


Trigers are the SQL codes that are automatically executed in response to any certain events on a particular table.
These are used to maintain the integrity of data.


**Exp-**
Let's say multiple new_joineers are coming in a team so manager wants to send a Welcome_Mail to all of them.
As these joiners are multiple so we can use a triger here like whenever any new record inserted in a table so one mail will sent to him automatically.


Triggers are store procedure/program, which are automatically executed when some event occurs

Event can be any of the following:
 DML
 DDL
 Data base operation-(Logon Logoff Startup Startdown)








**Syntax-**

Create(or replace) Trigger Trigger_Name
[Before | After]
[Insert | Update | Delete]
on [Table_Name]
[for each row | for each column]
Declare                        (Declare part is optional)
